```python
import json, random, csv, shutil
from pathlib import Path

N = 5000                   
MIN_LEN, MAX_LEN = 5, 40   
ONE_CAPTION_PER_IMAGE = True
USE_SYMLINKS = True        

ROOT = Path('~/projects/text2image').expanduser()
COCO_IMG = ROOT/'data/coco_val/val2017'
COCO_CAP = ROOT/'data/coco_val/annotations/captions_val2017.json'
SMALL = ROOT/'data/coco_small'
SMALL_IMG = SMALL/'images'
pairs_csv = SMALL/'pairs.csv'

with open(COCO_CAP,'r') as f:
    data = json.load(f)

img_by_id = {im['id']: im for im in data['images']}
cap_by_img = {}

for a in data['annotations']:
    img_id = a['image_id']
    if ONE_CAPTION_PER_IMAGE and img_id in cap_by_img:
        continue
    cap = a['caption'].strip()
    toks = cap.split()
    if MIN_LEN <= len(toks) <= MAX_LEN:
        cap_by_img[img_id] = cap

items = []
for img_id, cap in cap_by_img.items():
    fname = img_by_id[img_id]['file_name']
    items.append((fname, cap))

random.seed(123)
random.shuffle(items)
items = items[:N]

with open(pairs_csv, 'w', newline='') as f:
    w = csv.writer(f)
    w.writerow(['image_path','caption'])
    for fname, cap in items:
        src = COCO_IMG/fname
        dst = SMALL_IMG/fname
        if not dst.exists():
            if USE_SYMLINKS:
                try:
                    dst.symlink_to(src)
                except FileExistsError:
                    pass
            else:
                shutil.copy2(src, dst)
        w.writerow([str(dst), cap])

print("Wrote", len(items), "pairs to", pairs_csv)
!head -n 5 ~/projects/text2image/data/coco_small/pairs.csv
```

```python
from pathlib import Path
import torch
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt

# Output folder
ROOT = Path('~/projects/text2image').expanduser()
OUT = ROOT/'outputs/samples'
OUT.mkdir(parents=True, exist_ok=True)

# Five example prompts (you may edit these)
prompts = [
    "A small red bicycle parked next to a brick wall, late afternoon",
    "Two golden retriever puppies playing in tall green grass",
    "A bustling street market at night with colorful neon signs",
    "A wooden dining table with a bowl of fresh oranges, soft natural light",
    "A surfer riding a giant wave under a dramatic cloudy sky"
]

device = "cuda" if torch.cuda.is_available() else "cpu"
dtype = torch.float16 if device == "cuda" else torch.float32

# Load pipeline
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=dtype
).to(device)

# Memory-friendly settings (safe on most clusters)
pipe.enable_attention_slicing()
try:
    pipe.enable_xformers_memory_efficient_attention()
except Exception:
    pass

# Generate and save
images = []
for i, p in enumerate(prompts, 1):
    result = pipe(
        p,
        num_inference_steps=30,   # adjust if you want faster/slower
        guidance_scale=7.5,
        height=512,
        width=512
    )
    img = result.images[0]
    img_path = OUT / f"sample_{i:02d}.png"
    img.save(img_path)
    images.append((p, img_path))

# Show results inline
for p, path in images:
    plt.figure(figsize=(5,5))
    plt.imshow(plt.imread(path))
    plt.axis("off")
    plt.title(p)
    plt.show()

print("Saved files:")
!ls -1 ~/projects/text2image/outputs/samples/
```

```python
from pathlib import Path
import pandas as pd
import numpy as np
import torch
from transformers import CLIPTokenizer, CLIPTextModel
from tqdm import tqdm

ROOT = Path('~/projects/text2image').expanduser()
PAIRS = ROOT / 'data/coco_small/pairs.csv'
OUT = ROOT / 'data/coco_small/text_clip_embeddings.npz'

# Load captions
df = pd.read_csv(PAIRS)
captions = df['caption'].tolist()

# Load CLIP text encoder + tokenizer
device = "cuda" if torch.cuda.is_available() else "cpu"
tokenizer = CLIPTokenizer.from_pretrained("openai/clip-vit-base-patch32")
text_model = CLIPTextModel.from_pretrained("openai/clip-vit-base-patch32").to(device).eval()

# Encode in batches
embeds = []
bs = 64
with torch.no_grad():
    for i in tqdm(range(0, len(captions), bs)):
        batch_caps = captions[i:i+bs]
        enc = tokenizer(batch_caps, padding=True, truncation=True, return_tensors="pt").to(device)
        out = text_model(**enc).pooler_output         # shape [B, 512]
        embeds.append(out.detach().cpu().numpy())

embeds = np.concatenate(embeds, axis=0)               # [N, 512]
np.savez_compressed(OUT, captions=np.array(captions, dtype=object), clip_text_emb=embeds)

print("Saved:", OUT)
print("Embeddings shape:", embeds.shape)
print("Example caption:", captions[0])
print("First vector (first 5 dims):", embeds[0][:5])
```
